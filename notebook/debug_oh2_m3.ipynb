```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../csv/20230919_091610.csv')
df
```

```python
df['Expected Result'] = df['Expected Result'].replace(np.nan, "Missing")
df['Actual Result'] = df['Actual Result'].replace(np.nan, "Missing")
```

```python
df_call = df[df['Criterion Name'] == "OH2 handle"]
df_call
```

```python
df_call_fail = df_call[df_call['Test Result NR'] == "Fail"]
df_call_fail
```

## Yes Label

```python
df_call_fail_yes = df_call_fail[df_call_fail['Expected Result'] == "Yes"]
df_call_fail_yes
```

### Yes > No

```python
df_call_fail_yes_no = df_call_fail_yes[df_call_fail_yes['Actual Result'] == "No"]
df_call_fail_yes_no
```

```python
df_call_fail_yes_no['Actual Reason'].value_counts()
```

### Yes -> N/A

```python
df_call_fail_yes_na = df_call_fail_yes[df_call_fail_yes['Actual Result'] == "Missing"]
df_call_fail_yes_na
```

```python
df_call_fail_yes_na['Expected Reason'].value_counts()
```

## No label

```python
df_call_fail_no = df_call_fail[df_call_fail['Expected Result'] == "No"]
df_call_fail_no
```

### No -> Yes

```python
df_call_fail_no_yes = df_call_fail_no[df_call_fail_no['Actual Result'] == "Yes"]
df_call_fail_no_yes
```

```python
df_call_fail_no_yes['Expected Reason'].value_counts()
```

### No -> N/A

```python
df_call_fail_no_na = df_call_fail_no[df_call_fail_no['Actual Result'] == "Missing"]
df_call_fail_no_na
```

```python
df_call_fail_no_na['Expected Reason'].value_counts()
```

## N/A Label

```python
df_call_fail_na = df_call_fail[df_call_fail['Expected Result'] == "Missing"]
df_call_fail_na
```

### N/A -> Yes

```python
df_call_fail_na_yes = df_call_fail_na[df_call_fail_na['Actual Result'] == "Yes"]
df_call_fail_na_yes
```

```python
df_call_fail_na_yes['Actual Reason'].value_counts()
```

### N/A -> No

```python
df_call_fail_na_no = df_call_fail_na[df_call_fail_na['Actual Result'] == "No"]
df_call_fail_na_no
```
