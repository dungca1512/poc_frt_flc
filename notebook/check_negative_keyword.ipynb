```python
import json
```

```python
with open("../data/4_call_transcript.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
data
```

```python
with open("../keyword/Thiếu chuyên nghiệp.txt", "r", encoding="utf-8") as f:
    keywords = f.read()
keywords = keywords.split(",")
print(keywords)
```

```python
count_keyword = []
```

```python
for call in data:
    # print(call["content"])
    for content in call["content"]:
        count = 0
        for keyword in keywords:
            if content["channel"] == 2 and keyword in content["text"]:
                count += 1
    count_keyword.append({"name": call["name"], "count": count})
```
